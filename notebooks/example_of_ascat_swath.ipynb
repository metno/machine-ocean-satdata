```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
data_dir = "/lustre/storeB/project/IT/geout/machine-ocean/data_raw/metop/"
ascat_fn = 'ASCATL1SZR_20180329T145700Z_20180329T163558Z_epct_4695022c_P.nc'
data_ascat = xr.open_dataset(data_dir + ascat_fn)
```

```python
data_ascat['sigma0_trip_fore'].plot(vmax=10, vmin=-40)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
data_ascat['sigma0_trip_fore'].plot(
    ax=ax, 
    vmax=10, 
    vmin=-40,
    cbar_kwargs={"orientation": "horizontal", "shrink": 0.7, 'extend': 'both'},
)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_extent([-75, -65, 35, 45], crs=ccrs.PlateCarree())
data_ascat['sigma0_trip_fore'].plot(
    ax=ax, 
    vmax=10, 
    vmin=-40,
    cbar_kwargs={"orientation": "horizontal", "shrink": 0.5, 'extend': 'both'},
)
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

pioneer_lon = -70.8
pioneer_lat = 40
ax.scatter([pioneer_lon], [pioneer_lat],
         color='r',
         transform=ccrs.PlateCarree(),
         )

ax.text(
    pioneer_lon - 0.1, 
    pioneer_lat - 0.5, 
    'Pioneer',
    horizontalalignment='right',
    #color='r',
    fontsize=13,
    weight='bold',
    transform=ccrs.Geodetic()
    )
```
