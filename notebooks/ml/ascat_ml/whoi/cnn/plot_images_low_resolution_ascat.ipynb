### Plot low-resolution images that collocate with Irminger 

The file paths of the images have been identified in the notebook `check_resolution_ascat.ipynb`

```python
irminger_lat = 59.9337
irminger_lon = -39.47378
```

```python
data_dir = "/lustre/storeB/project/IT/geout/machine-ocean/data_raw/metop/"
ascat_fn = 'ASCATL1SZR_20200121T221800Z_20200122T000258Z_epct_dde0d061_P.nc'
data_ascat = xr.open_dataset(data_dir + ascat_fn)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
data_ascat['sigma0_trip_fore'].plot(vmax=0, vmin = -30)
ax.gridlines(crs=ccrs.PlateCarree(), linewidth=1, color='black', alpha=0.5, linestyle='--', draw_labels=True)
plt.scatter(irminger_lon, irminger_lat, color='red', s=50)
ax.set_extent([-60, -20, 50, 90])
```
