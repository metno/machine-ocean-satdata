```python
dict_f_usable_aft = {}
dict_f_usable_mid = {}
dict_f_usable_fore = {}
for buoy in ascat_dict.keys():
    print(buoy)
    try:
        dict_f_usable_aft[buoy] = 0
        dict_f_usable_mid[buoy] = 0
        dict_f_usable_fore[buoy] = 0
        for key, ascat_fn in ascat_dict[buoy]['nc_files'].items():
            data_ascat = xr.open_dataset(data_dir + ascat_fn)
            if 2 in data_ascat.f_usable_aft:
                dict_f_usable_aft[buoy] = dict_f_usable_aft[buoy] + 1
            if 2 in data_ascat.f_usable_mid:
                dict_f_usable_mid[buoy] = dict_f_usable_mid[buoy] + 1
            if 2 in data_ascat.f_usable_fore:
                dict_f_usable_fore[buoy] = dict_f_usable_fore[buoy] + 1
    except:
        print('No nc files found.')
    print('No. of images with bad data in aft: ', dict_f_usable_aft[buoy])
    print('No. of images with bad data in mid: ', dict_f_usable_mid[buoy])
    print('No. of images with bad data in fore: ', dict_f_usable_fore[buoy])
```
